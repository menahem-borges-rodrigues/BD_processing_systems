### Load to Hive

```python
%%file SPBD-P1-G07-hive-createload.hsql

-- Apaga tabela
drop table epa_hap_daily_summary;
drop table usa_states;

-- Cria tabela
CREATE TABLE epa_hap_daily_summary(
state_code             int,
county_code            int,
site_num               int,
parameter_code         int,
poc                    int,
latitude               float,
longitude              float,
datum                  CHAR(28),
parameter_name         CHAR(28),
sample_duration        CHAR(28),
pollutant_standard     float,
date_local             CHAR(28),
units_of_measure       CHAR(28),
event_type             CHAR(28),
observation_count      int,
observation_percent    float,
arithmetic_mean        float,
first_max_value        float,
first_max_hour         int,
aqi                    float,
method_code            int,
method_name            CHAR(28),
local_site_name        CHAR(28),
address                CHAR(28),
state_name             CHAR(28),
county_name            CHAR(28),
city_name              CHAR(28),
cbsa_name              CHAR(28),
date_of_last_change    CHAR(28))

ROW FORMAT DELIMITED FIELDS TERMINATED BY ',' -- separador
TBLPROPERTIES ("skip.header.line.count"="1");

-- Carrega ficheiro .csv
load data local inpath 'epa_hap_daily_summary-small.csv' into table epa_hap_daily_summary;

CREATE TABLE usa_states (
state                  CHAR(4),
name                   CHAR(28),
MinLat                 float,
MaxLat                 float,
MinLong                float,
MaxLong                float
)

ROW FORMAT DELIMITED FIELDS TERMINATED BY ',' -- separador
TBLPROPERTIES ("skip.header.line.count"="1");

load data local inpath 'usa_states.csv' into table usa_states;

-- Cria view para tirar datas nulas
drop view epa_hap;
create view if not exists epa_hap as select * from epa_hap_daily_summary where date_local is not NULL;
```

# Q1
### Which states have more/less monitors? (Rank states!).

```python
%%file SPBD-P1-G07-hive-Q1.hsql

SELECT state_name, count(distinct state_code, latitude, longitude) AS number_monitors
        FROM epa_hap
        GROUP BY state_name
        SORT BY number_monitors DESC;
```

```python
!hive -v -f SPBD-P1-G07-hive-Q1.hsql
```

# Q2
### Which counties have the best/worst air quality? (Rank counties considering pollutants’ level! ).

```python
%%file SPBD-P1-G07-hive-Q2.hsql

SELECT county_name, avg(arithmetic_mean) AS media_poluicao
        FROM epa_hap
        GROUP BY county_name
        ORDER BY media_poluicao DESC;
```

```python
!hive -v -f SPBD-P1-G07-hive-Q2.hsql
```

# Q3
### Which states have the best/worst air quality in each year? (Rank states per year considering pollutants’ level!).

```python
%%file SPBD-P1-G07-hive-Q3.hsql

SELECT t.data, t.state_name, avg(t.arithmetic_mean) AS media FROM (
SELECT substring(date_local,0,4) AS data, state_name, arithmetic_mean FROM epa_hap 
    ) t
    GROUP BY t.data, state_name
    ORDER BY t.data, media;
```

# Q4
### For each state, what is the average distance (in km) of the monitors in that state to the state center? For simplicity, assume that 1 degree of latitude or logitude equals to 111 km. (Monitor dispersion per state!)].

```python
%%file SPBD-P1-G07-hive-Q4.hsql

DROP table monitores;
CREATE temporary table monitores AS
    SELECT distinct state_name, latitude, longitude
            FROM epa_hap;

DROP table state_center;
CREATE temporary table state_center AS
    SELECT name, (MinLat+MaxLat)/2 AS lat_center, (MinLong+MaxLong)/2 AS lon_center 
            FROM usa_states;
    
DROP table full_data;    
CREATE temporary table full_data AS
    SELECT monitores.state_name, monitores.latitude, monitores.longitude, state_center.lat_center, state_center.lon_center
            FROM monitores 
            LEFT JOIN state_center
            ON monitores.state_name = state_center.name;
        
SELECT state_name, avg(sqrt(power((latitude-lat_center)*111,2)+power((longitude-lon_center)*111,2))) AS dist 
        FROM full_data 
        GROUP BY state_name
        ORDER BY dist desc;
```

# Q5
### How many sensors there are per quadrant (NW, NE, SE, SW) in each state? To answer this question, you should approximate each state’s area to a rectangle as defined in the file “usa_satates.csv”, and divide that area in 4 quadrants (NW, NE, SE, SW). (Count monitors per sate qudrant! )].

```python
%%file SPBD-P1-G07-hive-Q5.hsql

DROP table monitores;
CREATE temporary table monitores AS
    SELECT distinct state_name, latitude, longitude
            FROM epa_hap;
        
DROP table state_center;
CREATE temporary table state_center AS
    SELECT name, (MinLat+MaxLat)/2 AS lat_center, (MinLong+MaxLong)/2 AS lon_center 
            FROM usa_states;
        
DROP table full_data;    
CREATE temporary table full_data AS
    SELECT monitores.state_name, monitores.latitude, monitores.longitude, state_center.lat_center, state_center.lon_center
            FROM monitores 
            LEFT JOIN state_center
            ON monitores.state_name = state_center.name;
            
DROP table mon_quad;
CREATE temporary table mon_quad as
    SELECT state_name, 
    CASE WHEN latitude > lat_center and longitude > lon_center THEN 1 ELSE 0 END AS NE, 
    CASE WHEN latitude < lat_center and longitude > lon_center THEN 1 ELSE 0 END AS SE,
    CASE WHEN latitude<lat_center and longitude<lon_center THEN 1 ELSE 0 END AS SW,
    CASE WHEN latitude>lat_center and longitude<lon_center THEN 1 ELSE 0 END AS NW 
            FROM full_data;
    
SELECT state_name,SUM(NE) AS NE,SUM(SE) AS SE,SUM(SW) AS SW,SUM(NW) AS MW 
            FROM mon_quad GROUP BY state_name;
```
