```python
#Q.4) For each state, what is the average distance (in km) of the monitors in that state to the state center? 
#For simplicity, assume that 1 degree of latitude or logitude equals to 111 km. (Monitor dispersion per state!)
import pyspark
import math
from operator import add as sum
from time import process_time,perf_counter
t_start,r_start=process_time(),perf_counter()
sc = pyspark.SparkContext('local[*]')
try :
    data = sc.textFile('epa_hap_daily_summary-small.csv') #main data
    data_2= sc.textFile('usa_states.csv') #auxialiar data
    header,header2 = data.first(),data_2.first() #nomear primeira linha
    data = data.filter(lambda row: row != header) #extrair primeira linha 
    data_2 = data_2.filter(lambda row: row != header2) #extrair primeira linha 
    non_empty_lines = data.filter( lambda line : len(line) > 0 ) #extrair linhas vazias
    non_empty_lines2 = data_2.filter( lambda line : len(line) > 0 ) #extrair linhas vazias
    state_monitor=non_empty_lines.map( lambda line : line.split(',')).map(lambda v: ((v[5],v[6]),v[24])).reduceByKey(lambda a,b: a).map(lambda x: (x[1],x[0])) #criar map formato estado,monitor
    state_center=non_empty_lines2.map(lambda line:line.split(',')).map(lambda word:(word[1],((float(word[2])+float(word[3]))/2,(float(word[4])+float(word[5]))/2)))
    data_joint=state_monitor.join(state_center).map(lambda x: (x[0],(math.sqrt(((float(x[1][0][0])-x[1][1][0])*111)**2 + ((float(x[1][0][1])-x[1][1][1])*111)**2),1)))
    state_query = data_joint.reduceByKey( lambda a,b: (a[0] + b[0],a[1]+b[1])).map(lambda r: (r[0],r[1][0]/r[1][1])).sortBy(lambda s:s[1],ascending=False)
    for s,n in state_query.collect():
        print(s,n)
    t_stop,r_stop=process_time(),perf_counter()
    print(f'process time, real time {t_stop-t_start,r_stop-r_start}')
    sc.stop()
except Exception as inst:
    print(inst,type(inst))
    sc.stop()
```
